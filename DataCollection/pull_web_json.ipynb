```python
zipcode_df = {}
zipcode_df = pd.DataFrame(zipcode_df)
zipcode_df["zipcode"] = ''
for agegroup in responseJSON["zip_values"][0]["demographics"]["age"]:
    if agegroup["age_group"] == 'Unknown':
        continue
    zipcode_df[agegroup["age_group"]] = ''
for race in responseJSON["zip_values"][0]["demographics"]["race"]:
    zipcode_df[race["description"]] = ''
for gender in responseJSON["zip_values"][0]["demographics"]["gender"]:
    zipcode_df[gender["description"]] = ''
    
########################
zipcode_row = {}
for zipcode in responseJSON["zip_values"]:
    zipcode_row["zipcode"] = (zipcode["zip"]) #adds the zip code
    for agegroup in zipcode["demographics"]["age"]: #fill each age column with data
        if agegroup["age_group"] == 'Unknown':
            continue
        zipcode_row[agegroup["age_group"]] = (agegroup["count"]) #adds count to appropriate agegroup column
    for race in zipcode["demographics"]["race"]: 
        zipcode_row[race["description"]] = (race["count"]) #fills race columns with data
    for gender in zipcode["demographics"]["gender"]:
        zipcode_row[gender["description"]] = (gender["count"])
    zipcode_df = zipcode_df.append(zipcode_row,ignore_index=True)


    #zipcode_df.append(zipcode_row,ignore_index=True)
zipcode_df["zipcode"] = pd.to_numeric(zipcode_df["zipcode"])
zipcode_df.head()
```

```python
zipcode_df["county"] = ''
for index, row in zipcode_df.iterrows():
    try:
        row_county = zipcode_list.loc[zipcode_list["zip"] == row["zipcode"],"county_name"].iloc[0]
        zipcode_df.at[index,"county"] = row_county
    except IndexError:
        print(row["zipcode"],"zipcode not in database")
zipcode_df.head()
    
```

```python
for column in zipcode_df.columns: 
    if column == 'county':
        continue
    zipcode_df[column] = pd.to_numeric(zipcode_df[column])
county_df = zipcode_df.groupby("county").sum()
county_df.drop(columns=["zipcode"],inplace=True)
county_df.head()
```
